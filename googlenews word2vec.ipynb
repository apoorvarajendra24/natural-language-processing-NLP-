```python
import gensim
```

```python
from gensim.models import Word2Vec, KeyedVectors
```

```python
# Using Google's pretrained model to create vectors
model_path = r'E:\padhai ka saman\python\NLP\googlenews-vector-negative-300\GoogleNews-vectors-negative300.bin'
wv = KeyedVectors.load_word2vec_format(model_path, binary=True)
```

```python
# dense vector representation of the word King with 300 dimensions (300 feature representations, embeddiing size,N = 300)
wv['king']
```

```python
def word_embeddings(word):
    try:
        print(f'Word Embedding for {word}:\n{wv[word]}')
    except KeyError:
        print(f"{word} not in vocabulary")
```

```python
word_embeddings('babe')
```

```python
wv['king'].shape
```

```python
# Measuring word similarity (internally uses cosine similarity)
wv.similarity('King','Queen')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(wv['Honey'].reshape(1,-1),wv['Bees'].reshape(1,-1))
```

```python
cosine_similarity(wv['cheetah'].reshape(1,-1),wv['leopard'].reshape(1,-1))
```

```python
# Finding the most similar words
wv.most_similar('Public')
```

```python
# The famous equation king-man+woman
wv.most_similar(positive=['king', 'woman'], negative=['man'])
```

```python
# Odd one out
wv.doesnt_match(['tennis','cricket','basketball','volleyball','ludo'])
```

```python
# Cosine distance between two words
wv.distance('waffle','bread')
```

```python
# Ranking similar words
wv.most_similar_to_given('king', ['man', 'queen', 'apple'])
```
