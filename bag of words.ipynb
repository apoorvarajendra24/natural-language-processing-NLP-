```python
import pandas as pd
```

```python
messages = pd.read_csv("spam_classification/spam.csv",encoding="ISO-8859-1")
messages = messages[['v1','v2']]
messages.columns = ['labels','message']
messages
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
corpus = []
```

```python
for i in range(len(messages)):       # iterate thru length of the dataset messages
    review = re.sub('[^a-zA-z]', " ", messages['message'][i])       # replace with space if not anything of a-zA-Z    (most recommended)
    review = review.lower()       # lower case the sentence
    review = review.split()       # splits each word in the sentence
    words = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = " ".join(words)      # join them back
    corpus.append(review)
```

```python
corpus
```

```python
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer(max_features=500)         # most frequent top 500 words makes the features
```

```python
X = bow.fit_transform(corpus).toarray()
```
